```python
# ============================================
# LATIHAN ANALISIS DATASET TITANIC (PANDAS)
# ============================================

import pandas as pd
import numpy as np

# --------------------------------------------
# LOAD DATASET TITANIC DARI GITHUB
# --------------------------------------------
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Tampilkan 5 data pertama
print("=== 5 DATA PERTAMA ===")
print(df.head())
```

```python
# ============================================
# Q1. Jumlah baris dan kolom + nama kolom
# ============================================
print("\n=== Q1 ===")
print("Jumlah baris dan kolom:", df.shape)
```

```python
# ============================================
# Q2. Missing Values per Kolom
# ============================================
print("\n=== Q2 ===")

missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Jumlah Missing': missing_count,
    'Persentase (%)': missing_percent.round(2)
})

print(missing_df[missing_df['Jumlah Missing'] > 0])
```

```python
# ============================================
# Q3. Persen Penumpang Selamat (Survived=1)
# Gunakan NumPy / Pandas
# ============================================
print("\n=== Q3 ===")

# Cara Pandas
survival_rate = df['Survived'].mean() * 100
print("Persentase selamat (Pandas):", round(survival_rate, 2), "%")

# Cara NumPy
survival_rate_np = np.mean(df['Survived']) * 100
print("Persentase selamat (NumPy):", round(survival_rate_np, 2), "%")
```

```python
# ============================================
# Q4. Filter wanita kelas 1
# Jumlah & rata-rata usia
# ============================================
print("\n=== Q4 ===")

female_class1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

print("Jumlah penumpang wanita kelas 1:", len(female_class1))
print("Rata-rata usia:", round(female_class1['Age'].mean(), 2))
```

```python
# ============================================
# Q5. Groupby tingkat keselamatan per kelas
# ============================================
print("\n=== Q5 ===")

survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100

print("Tingkat keselamatan per kelas:")
print(survival_by_class.round(2))

kelas_tertinggi = survival_by_class.idxmax()
print("Kelas dengan tingkat keselamatan tertinggi: Pclass", kelas_tertinggi)
```
